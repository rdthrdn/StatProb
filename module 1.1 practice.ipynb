SOAL 1 : Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

```python
# Membuat list dari tinggi badan siswa kelas A
tinggi = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung mean, median, dan modus dari list
import statistics
mean = statistics.mean(tinggi)
median = statistics.median(tinggi)
modus = statistics.mode(tinggi)

# Mencetak hasil perhitungan
print("Mean tinggi badan siswa kelas A adalah", mean, "cm")
print("Median tinggi badan siswa kelas A adalah", median, "cm")
print("Modus tinggi badan siswa kelas A adalah", modus, "cm")

# Menggambar histogram dari list
import matplotlib.pyplot as plt
plt.hist(tinggi, bins=10)
plt.xlabel("Tinggi badan (cm)")
plt.ylabel("Frekuensi")
plt.title("Histogram tinggi badan siswa kelas A")
plt.show()
```

SOAL 2 : Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A: 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

```python
# Membuat list dari nilai UAS pada mata pelajaran matematika di kelas A
nilai = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]

# Menghitung mean dari list
import statistics
mean = statistics.mean(nilai)

# Mencetak hasil perhitungan
print("Mean nilai UAS pada mata pelajaran matematika di kelas A adalah", mean)

# Menggambar histogram dari list
import matplotlib.pyplot as plt
plt.hist(nilai, bins=10)
plt.xlabel("Nilai UAS")
plt.ylabel("Frekuensi")
plt.title("Histogram nilai UAS pada mata pelajaran matematika di kelas A")
plt.show()
```
